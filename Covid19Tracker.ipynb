# COVID-19 

```
Symptoms = ["cough","breathing","fever"]
#Symptoms = [ele for ele in syptoms]
Testing = ["P","N"]
```

```
#making function 
def make_agent(Symptoms,Testing):
  return [Symptoms,Testing]
agent_one = make_agent(Symptoms[1],Testing[0])
print(agent_one)
```

```
import random

def Testing_population(N):
  population =[]
  for i in range(N):
    v=random.choice(Symptoms)
    p=random.randint(0,1)
    agent =make_agent(v,Testing[p])
    population.append(agent)
  return population


results =Testing_population(100)
print(results)
```

```
print('P' in results[1])
positive_cases= []
for pos in results:
    if 'P' in pos:
        positive_cases.append(pos)
    print(pos)
len(positive_cases)
```

```
Active_Cases = len(positive_cases)
Tested_Cases = len(results)
Negative_Cases=Tested_Cases-Active_Cases
print(" Total Test Cases",Tested_Cases," Positive Cases",Active_Cases,"Negative Cases",Negative_Cases)
```

```
from matplotlib import pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ['Corona Test Positive', 'Corona Test Negative']
Patients = [Active_Cases,Negative_Cases]
ax.pie(Patients, labels = langs,autopct='%1.2f%%')
plt.show()
```

```
import matplotlib.pyplot as plt 
  
# line 1 points 
x1 = [0,Tested_Cases] 
y1 = [0,Active_Cases] 
# plotting the line 1 points  
plt.plot(x1, y1, label = "Positive Test Cases = "+str(Active_Cases)) 
  
# line 2 points 
x2 = [0,Tested_Cases] 
y2 = [0,Negative_Cases] 
# plotting the line 2 points  
plt.plot(x2, y2, label = "Negative Test Cases = "+str(Negative_Cases)) 
  
# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 
# giving a title to my graph 
plt.title('COVID-19 Graphical Representation Of c!') 
  
# show a legend on the plot 
plt.legend() 
  
# function to show the plot 
plt.show() 
```
